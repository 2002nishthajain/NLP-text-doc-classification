```python
import pandas as pd
df = pd.read_csv(r'Medical_dataset.csv', encoding='latin-1')
df.head()
```

```python
import pandas as pd
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words('english'))
punctuations = set(string.punctuation)

lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    text = text.lower()
    tokens = word_tokenize(text)
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words and token not in punctuations and token.isalpha()]
    return ' '.join(tokens)  


df['description'] = df['description'].apply(preprocess_text)


df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

# Fit and transform the 'description' column
X = vectorizer.fit_transform(df['description'])

# Find the highest frequency words in each row
max_freq_words = []
for i, description in enumerate(df['description']):
    row = X[i]
    word_idx = row.argmax()
    max_freq_word = list(vectorizer.vocabulary_.keys())[list(vectorizer.vocabulary_.values()).index(word_idx)]
    max_freq_words.append(max_freq_word)

# Add a new 'title' column to the DataFrame with the maximum frequency words
df['Title'] = max_freq_words
```

```python
df['category'] = "Medical"
df['date'] = '2023-10-08'
df['date'] = pd.to_datetime(df['date'])
```
